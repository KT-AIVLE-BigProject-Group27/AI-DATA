```python
import pandas as pd
import json
```

```python
train_data = pd.read_csv('/content/drive/MyDrive/생성데이터/train_exa.csv')
validation_data = pd.read_csv('/content/drive/MyDrive/생성데이터/validation_exa.csv')
```

```python
import pandas as pd
import json

# train_data와 validation_data를 딕셔너리로 변환
train_dict = train_data.to_dict(orient='records')
validation_dict = validation_data.to_dict(orient='records')

# JSON 파일로 저장
with open('train_data.json', 'w', encoding='utf-8') as train_file:
    json.dump(train_dict, train_file, ensure_ascii=False, indent=4)

with open('validation_data.json', 'w', encoding='utf-8') as val_file:
    json.dump(validation_dict, val_file, ensure_ascii=False, indent=4)

print("JSON 파일로 저장 완료!")
```
